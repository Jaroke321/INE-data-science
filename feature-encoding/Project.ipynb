## Nominal/Multi-category/Discrete Encoding

```python
fpath = 'datasets/nominal_data.csv'       # Enter the local filepath to your nominal_data.csv file
df_nom = pd.read_csv(fpath)
df_nom.head()
```

```python
# Dummy encode nom_1
# Fill in your code here
df_nom_dummies = pd.get_dummies(df_nom, columns=['nom_1'], drop_first=True)
df_nom_dummies.head()
```

```python
# Check that the number of dummy columns is 1 fewer than the number
# of unique categories in nom_1
print('Unique values in nom_1: {}'.format(df_nom.nom_1.unique().shape[0]))
print('Unique values in nom_1_dummies: {}'.format(df_nom_dummies.columns.str.startswith('nom_1').sum()))
```

## Ordinal Encoding

```python
fpath = 'datasets/ordinal_data.csv'      # Enter the local filepath to your ordinal_data.csv file
df_ord = pd.read_csv(fpath)
df_ord.head()
```

```python
# Encode ord_2 into a feature called 'ord_2_mapped'
# Fill in your code here
df_ord.ord_2.unique()
df_ord['ord_2_mapped'] = df_ord.ord_2.map({'Freezing': 0, 'Cold': 1, 'Warm': 2, 'Hot': 3, 'Boiling Hot': 4, 'Lava Hot': 5})
df_ord.head()
```

```python
# Encode ord_4 into a feature called 'ord_4_mapped'
# Fill in your code here
temp = pd.DataFrame(df_ord.ord_4.unique()).sort_values(by=0).reset_index()
ord_map = { a:b for a, b in zip(temp[0], temp.index) }

df_ord['ord_4_mapping'] = df_ord.ord_4.map(ord_map)
df_ord.head()
```

## Cyclical Encoding

```python
fpath = 'datasets/cyclical_data.csv'    # Enter the local filepath to your cyclical_data.csv file
df_cyclical = pd.read_csv(fpath)
df_cyclical.head()
```

```python
# Encode cyclical columm 'month'
# Fill in your code here
df_month_dummies = pd.get_dummies(df_cyclical.month, drop_first=True)
df_month_dummies.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
# Check for collinearity in your dummy features
# Fill in your code here
X = add_constant(df_month_dummies)
vif = pd.DataFrame()
vif['feature'] = X.columns

vif['collinearity'] = [ variance_inflation_factor(X.values, i) 
                      for i in range(X.shape[1])]

vif
```

```python
# Join the month dummies back to the day column in the cyclical data
# Fill in your code here
df_cyclical = pd.concat([df_cyclical[['day']], df_month_dummies], axis=1)
df_cyclical.head()
```
